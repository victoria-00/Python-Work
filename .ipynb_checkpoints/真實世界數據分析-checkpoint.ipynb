```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("https://quality.data.gov.tw/dq_download_csv.php?nid=6288&md5_url=0563418ce8c623cdd5c5d918eea06d62")
df.head()
```

## 擷取資料中屬於中文圖書收藏的欄位

```python
dfs = df.iloc[:, 0:11]
dfs
```

## 將中文圖書收藏數加總

```python
dfs["中文圖書收藏總數"]=dfs.iloc[:, 2:11].sum(axis=1)
dfs.head()
```

## 排序

```python
dfs.sort_values(by="中文圖書收藏總數", ascending=False).head()
```
