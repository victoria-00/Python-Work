```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import pandas_datareader.data as web
```

```python
df = web.DataReader("AAPL","yahoo",start="2012-09-01")
```

```python
df.head()
```

```python
df.tail()
```

### 取出關盤價

```python
P = df["Close"]
```

### 計算每一期之間的報酬率

```python
P.diff()
```

### 跟第一期的資料比較，計算每一期之間的報酬率

```python
r = P.diff() / P
r.plot()
```

### 看前一百天開始的紀錄

```python
r[-100:].plot()
```

### 移動平均:透過window=20參數可以將前20天的累計值計算出來

```python
P.plot()
P.rolling(window=20).mean().plot()
P.rolling(window=60).mean().plot()
```
