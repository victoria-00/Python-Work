# 自己作dataframe

```python
mydata = np.random.randn(4,3)
```

```python
mydata
```

```python
list("ABCDE") #把一個字串變成list : list("字串")
```

```python
df1 = pd.DataFrame(mydata, columns=list("ABC"))

#自訂dataframe:新dataframe=pd.DataFrame(array, columns=項目list)
```

```python
df1
```

```python
df2 = pd.DataFrame(np.random.randn(3,3), columns=list("ABC"))
```

```python
df2
```

```python
df3 = pd.concat([df1,df2], axis=0, )  #pd.concat:df合併

#新df=pd.concat(dataframe的list, axis=0) #axis控制合併的方向─ 0是上下(列方向)，1是左右(行方向)
```

```python
df3
```

```python
df3.index=range(7)  #更改df的index :df名稱.index=要給的list
```

```python
df3
```

```python
df4=pd.concat([df1,df2],axis=1)
```

```python
df4
```

```python
df = pd.read_csv("http://bit.ly/uforeports")  #csv也可以是網址
```

```python
df.head()
```

```python
df_state = df.groupby("State").count()

#依項目集合值: dataframe名稱.groupby(項目)
#計算次數: dataframe名稱.count()
```

```python
df_state
```

```python
df_state.sort_values(by="Time", ascending=False)
```

```python
df_state
```

```python
df_state.sort_values(by="Time", ascending=False, inplace=True)

#sort不產生新的df的方法 : inplace=True
```

```python
df_state
```

```python
df_state.head(10)
```

```python
df_state[:10].Time.plot(kind='bar') #畫長條圖:df名稱.項目.plot(kind='bar')
```
