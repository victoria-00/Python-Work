## map and  filiter

```python
items = [500, 30, 2000]
c = 30.8594353
```

```python
L=[]               #先設一個新的List
for i in items:    #對於items中第i項
    twd = c*i      
    L.append(twd)  #append():加入括號內的東西 -> L.append(twd): 在L這個List中加入twd
print(L)           #印出L
```

```python
def exchange(twd):
    return twd*c
```

```python
exchange(500)
```

```python
list(map(exchange, items))  #map:套用函數
```

```python
A = list(map(exchange, items))
A
```

## lambdas 救急函數

```python
list(map(lambda p:p*c,items))
```

## list comprehesion

```python
[p*c for p in items]
```

## filter

```python
grades = [50, 59, 65, 80, 100, 35, 40, 88]
```

```python
#過濾某個條件之下

def pg(g):
    return g>=60
```

```python
pg(50)
```

```python
pg(88)
```

```python
list(filter(pg, grades)) #利用filter過濾掉grades中不符合條件pg的項
```

```python
list(filter(lambda g:g>=60, grades))
```

## print的格式化

```python
name = "宜家"
age = 19
print("您好,我是"+name+",今年"+str(age)+"歲。")
```

### method 01

```python
me = "您好,我是%s,今年%d歲" % (name,age) #字串: %s , 數字: %d
```

```python
print(me)
```

### method 02 : format

```python
me = "您好,我是{},今年{}歲".format(name,age)
print(me)
```

### method 03 : f-string

```python
me = f"您好,我是{name},今年{age}歲"
print(me)
```

## 更完整print控制

```python
egg= "ヽ(✿ﾟ▽ﾟ)ノ"

spam = f"■{egg}■"
print(spam)
```

### round 1

```python
spam=f"■{egg:20s}■"
print(spam)
```

### round 2

```python
spam=f"■{egg:>20s}■"
print(spam)
```

### round 3

```python
spam=f"■{egg:^20s}■"
print(spam)
```

```python
level = 6

for n in range(level):
    s = "*"*(2*n+1)
    print(f"{s:^21s}")
```

```python
spam = f"■{egg:*^20s}■"

print(spam)
```
